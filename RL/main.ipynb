```python
# Импорт numpy
import numpy as np
import matplotlib.pyplot as plt
```

```python
from system import System
from controller import (Controller, ConstantController, TimeBasedController, 
                       PDController, PIDController, CompositeController)
```

```python
# 3. PDController - пропорционально-дифференциальный контроллер
pd_controller = PDController(kp=0.5, kd=0.3, target=0.0)
system_pd = System(x0=np.array([2, 1]), controller=pd_controller)

N = 5000
for i in range(N):
    system_pd.update()

history_pd = system_pd.get_history()
time_pd = system_pd.get_time_history()

plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history_pd[:, 0], history_pd[:, 1])
plt.plot(system_pd.x0[0], system_pd.x0[1], 'ro', label='start')
plt.xlabel('Позиция')
plt.ylabel('Скорость')
plt.title('PDController (kp=0.5, kd=0.3)')
plt.grid(True)
plt.axhline(0, color='k', linestyle='--', alpha=0.3)
plt.axvline(0, color='k', linestyle='--', alpha=0.3)
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(time_pd, history_pd[:, 0])
plt.xlabel('Время')
plt.ylabel('Позиция')
plt.title('Позиция во времени')
plt.grid(True)
plt.axhline(0, color='r', linestyle='--', alpha=0.5, label='target')
plt.legend()

plt.tight_layout()
plt.show()
```
